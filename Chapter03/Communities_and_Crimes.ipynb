```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = df.replace('?', np.nan)
df.isnull().sum()
```

```python
for column in df.columns:
    print(column, end=' ')
```

```python
missing_val_count = df.isnull().sum()

for column in df.columns:
    print(column, missing_val_count.loc[column])
```

```python
value_count_df = df['state'].value_counts()
f, ax = plt.subplots(figsize=(20, 10))
value_count_df.plot.bar()
```

```python
value_count_df = df['state'].value_counts()
f, ax = plt.subplots(figsize=(20, 10))

value_count_df.plot.pie()
plt.show()
```

```python
def plot_hist(column, n_bins=100):
    f, ax = plt.subplots(figsize=(20, 10))
    df[column].hist(bins=n_bins)
    plt.show()
```

```python
plot_hist('population', n_bins=200)
```

```python
plot_hist('householdsize', n_bins=200)
```

```python
df = df.rename(columns={'racepctblack': 'racePctBlack'})

races = ['Black', 'White', 'Asian', 'Hisp']

for race in races:
    df['raceCnt' + race] = (df['population'] * df['racePct' + race]).astype(int)
    

df[['population'] + ['racePct' + race for race in races] + ['raceCnt' + race for race in races]].head()
```

```python
group_state_df = df.groupby('state')
group_state_df.sum()[['raceCnt' + race for race in races]]
```

```python
def plot_crime_box(columns):
    filtered_df = df[columns]
    
    f, ax = plt.subplots(figsize=(20, 10))
    sns.boxplot(filtered_df)
    ax.set_xticklabels(columns)
    
    plt.show()
```

```python
df = df.replace('?', np.nan)
plot_crime_box(['burglPerPop','larcPerPop','autoTheftPerPop','arsonsPerPop','nonViolPerPop'])
```

```python
filtered_df = df[['burglPerPop','larcPerPop','autoTheftPerPop','arsonsPerPop','nonViolPerPop']].astype(float)
filtered_df
```

```python
for column in filtered_df.columns:
    print(filtered_df[column].dtype)
```

```python
sns.boxplot(data=filtered_df)
```

```python
feature_columns = ['PctPopUnderPov', 'PctLess9thGrade', 'PctUnemployed', 'ViolentCrimesPerPop', 'nonViolPerPop']
df = df[feature_columns]

f, ax = plt.subplots(figsize=(13, 10))
sns.heatmap(df.dropna().astype(float).corr(), center=0, annot=True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)


plt.show()
```
