```python
import math
import numpy as np
```

### Calculate distance between Earth and sun

```python
def earth_sun_distance():
    # Semi-major axis between earth and sun
    A = 149600000
    # Eccentricity of earth
    E = 0.0167
    l = []
    for i in range(365):
        theta = (2 * math.pi * i) / 365.25
        r = A*(1 - E**2)/ (1 - (E * math.cos(theta)))
        l.append(r)
    
    return l
```

### Calculate time taken

```python
print("Earth sun distance:", "\n", l)
l = earth_sun_distance()


# Calculating the time taken
S = 299792
t = np.divide(l, S)

sunny = np.asarray(list(zip(l, t)))
print("Earth sun distance: ", sunny)

```
