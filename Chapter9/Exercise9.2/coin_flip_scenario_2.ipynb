```python
# coin_clip_scenario_2.py
# import the module
import random
import matplotlib.pyplot as plt

# set the sample_size or coin flips you what to run
sample_size = 20000

# initalize the variables required for our loop
# 1 = heads and 0 = tails
num_of_heads = 0
heads_list = []
trials_list = []
freq_list = []
# create a for loop and collect the results in a list
for i in range(1,sample_size+1):
    result = random.randint(0, 1)
    if result == 1:
        num_of_heads += 1
    avg_of_heads = float(num_of_heads) / i
    heads_list.append(num_of_heads)
    trials_list.append(i)
    freq_list.append(avg_of_heads)

# print the results
print(f'Results: {num_of_heads} heads out of {sample_size} flips. Average number of heads is {avg_of_heads}')

#create a simple line graph to show our results over time
plt.plot(trials_list, freq_list)
plt.ylabel('Sample Average')
plt.xlabel('Sample Size')
plt.hlines(0.50,0,sample_size,linestyles='dashed')
plt.show()
```
